###  *Nulls values cannot be filled in because we do not have accurate information or records to input and can affect the results so they have to be removed for accurate data analysis. The __.isnull__ function is used to drop null values.*

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df=df.rename(columns={'fixed acidity':'fixed_acidity','volatile acidity':'volatile_acidity','citric acid':'citric_acid','residual sugar':'residual_sugar','free sulfur dioxide':'free_sulfur_dioxide','total sulfur dioxide':'total_sulfur_dioxide'})
```

```python
df
```

```python
df.info()
```

```python
df.isnull()
```

```python
dfFill=df.mean()
```

```python
dfFill
```

```python
df=df.fillna(dfFill).drop_duplicates()
# Fills the missing values in the df with the values from dfFill, which contains the mean of each column in df and removes any duplicate rows from df.
```
